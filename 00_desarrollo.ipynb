```python
def extract_urls(link):
    url_lists = []
    
    # Vamos al link
    response = requests.get(link)
    urls = re.findall('/en/articles/archive/magic-story/.+', response.text)
    for url in urls:
        url_link = re.search('articles(.*)> ', url).group(1)[:-1]
        url_lists.append(url_link)
    
    return url_lists

```

```python
def get_text(link):
    # Vamos al link
    r = requests.get(link)
    # print(r.text)
    soup = BeautifulSoup(r.text, 'html.parser')
    # print(soup)
    # Buscamos los parrafos
    cuerpo = soup.find(id='content-detail-page-of-an-article')
    parrafos = cuerpo.find_all('p')
    # Agarramos solo el texto de cada parrafo y le sacamos \r y \n que definen nuevas lineas
    parrafos = [p.text.replace("\r","").replace("\n","").replace('\"', "") for p in parrafos]
    
    init = '<p>'
    close = '</p>'
    res = ''

# using loop to add string followed by delim 
    for ele in parrafos:
        res += init + str(ele) + close
  
    return res
```
